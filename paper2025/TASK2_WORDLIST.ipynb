```python
word_list = ["apple","window","bend","paper","thought"]
```

```python
# Task 2.1

# ask the user to input a new word
# take the new word as input
# convert the new word to lower case
# store the new word at the end of the list in a new space
# no need to consider any validation for the new word

new_word = input("Please enter a new word: ")
new_word = new_word.lower()
word_list.append(new_word)
print(word_list)
```

```python
# Task 2.2

# search the list to find words that have 5 or more letters
# count and output the number of words that have five or more letters

count = 0
for word in word_list:
    if len(word) >= 5:
        count += 1

# output message
print(f"Number of words with 5 or more letters: {count}")
```

```python
# Task 2.3

# search the list to find words that begin and end with the same letter
# count and output the number of words that begin and end with the same letter

count = 0
word_start_end_same = []
for word in word_list:
    if word[0] == word[-1]:
        count += 1
        word_start_end_same.append(word)

# output message
print(f"Number of words that begin and end with the same letter: {count}")
for word in word_start_end_same:
    print(word)
```

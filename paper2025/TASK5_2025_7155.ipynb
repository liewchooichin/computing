## Task 5

A company is creating a system to calculate, record and output different information relating to customer sales.

````python
# Task 5.1

# function total_cost(cost)
# sales tax is 9%
# calculating the total cost by adding the sales tax of 9%
# return the total cost

def total_cost(cost):
    sales_tax = cost * 0.09
    total_cost = cost + sales_tax
    return total_cost
````

````python
# Task 5.2

# A discount may be applied to a customer's purchase.

# Write a function discount(cost):
# deducts a discount of 5% from the total cost if the total cost is between
# $50 (inclusive) and $100 (exclusive).
# deducts a disount of 10% from the total cost if the total cost is
# $100 or greater.
# return the total cost with or without the discount as appropriate
# Must use the total_cost() function to calculate the total cost for the sale.

def discount(cost):
    if cost >= 50 and cost < 100:
        cost = cost * 0.95
    elif cost >= 100:
        cost = cost * 0.9
    return cost
````

````python
# Task 5.3

# A customer receives reward points on a purchase.

# Write a function reward_points(total_purchase):
# Calculates the number of reward points received for the purchase.
# A customer receives 3 reward points for each whole dollar($) spent.
# returns the number of reward points received.

# use math.floor()
import math

def reward_points(total_purchase):
    reward_points = math.floor(total_purchase)
    reward_points = reward_points * 3
    return reward_points
````

````python
# Task 5.4

# A customer may receive a voucher code that can be used
# for future purchase.

# Write a function voucher(total_cost, first_name):
# creates a voucher code that is the first three letters of the customer's name
# then the string "05PERCENT", if the total cost is between $25 (exclusive)
# and $50 (inclusive).
# creates a voucher code that is the first three letters of the customer's name
# then the string "10PERCENT", if the total cost is greater than $50.
# returns the voucher code or None as appropriate.

def voucher(total_cost, first_name):
    if total_cost > 25 and total_cost <= 50:
        voucher_code = first_name[:3] + "05PERCENT"
    elif total_cost > 50:
        voucher_code = first_name[:3] + "10PERCENT"
    else:
        voucher_code = None
    return voucher_code
````

````python
# Task 5.5

# The company wants an interface for the system.

# Create an interface with input() that:
# takes the first name of the customer as input,
# takes the cost of the sale as input,
# outputs a receipt for the customer, using the correct functions
# from the functions created above,
# the receipt should show the title "Receipt"
# the total cost of the sale (to 2 decimal place),
# the discounted cost of the sale (to 2 decimal place),
# the reward points received,
# the voucher code created.
# If no voucher code is created, output the text "You need to spend
# over $25 for a voucher code".
# If any voucher code is created, write the voucher code to a file
# named vouchercode.txt.

first_name = input("Enter your first name: ")
amount = float(input("Enter the cost of the sale: "))
amount_with_tax = total_cost(amount)
amount_after_discount = discount(amount)
reward_points_earned = reward_points(amount_after_discount)
voucher_code_applied = voucher(amount_after_discount, first_name)

# print the receipt
print("Receipt")
print(f"{'Amount with 9% tax: ':>28}{amount_with_tax:.2f}")
print(f"{'Amount after discounted: ':>28}{amount_after_discount:.2f}")
print(f"{'Reward points earned: ':>28}{reward_points_earned}")
if voucher_code_applied is None:
    print("You need to spend over $25 for a voucher code.")
else:
    print(f"{'Voucher code: ':>20}{voucher_code_applied}")

# write the file vouchercode.txt
if voucher_code_applied is not None:
    with open("vouchercode.txt", "w") as f:
        f.write(voucher_code_applied)
````

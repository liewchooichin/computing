The proram should put children into groups using the following rules:

- group 1  -- if the first letter of the first name is between A and M (inclusive) and they are older than 10 years.
- group 2 -- if the first letter of the first name is between N and Z (inclusive) and they are older than 10 years.
- group 3 -- if they are 10 years or younger.

- The program allows a child's first name and age to be entered.
- It then stores the child's firstname in the correct group list.
- The program loops until the user does not want to enter any more children's name or their ages.
- It then outputs the total number of children, as well as the names of the children in each group.


```python
group_1 = []
group_2 = []
group_3 = []
flag = True
count = 0
while flag:
    first_name = input("Please enter the child's name: ").upper()
    first_letter = first_name[0]
    age = input("Please enter the child's age: ")
    age = int(age)
    if age > 10:
      if first_letter >= "A" and first_letter <= "M" :
        group_1.append(first_name)
      elif first_letter >= "M" and first_letter <= "Z":
        group_2.append(first_name)
    elif age <= 10:
        group_3.append(first_name)
    count += 1
    more = input("Do you have another child to enter, Y or N?: ").upper()
    if more == "N":
        flag = False

# Print the output
print(f"You have entered the names of {count} children")
print("The members of group 1 are")
for i in group_1:
  print(i)
print("The members of group 2 are")
for i in group_2:
  print(i)
print("The members of group 3 are")
for i in group_3:
  print(i)


```
